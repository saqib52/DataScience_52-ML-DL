**Imputing methods for missing values**

```python
# finding out the missing value percentage
df.isnull().sum()/ len(df) * 100
```

```python
#1 drop the deck column from dataframe bcz have 77% missing values
df.drop('deck', axis= 1, inplace = True)
# df = df.drop(columns=['deck'])

# Without axis=1, Pandas would try to drop a row named 'deck'.
# inplace, Modifies the original DataFrame directly.
```

```python
df.columns
```

```python
# 2 filing missing values of age column by mean/median/mode of age
df['age'].mean()
df['age'].median()
# mode() hamesha list (Series) return karta hai → [0] se pehli value lo
# [1] tabhi lo jab confirm ho ke 2nd mode bhi exist karta hai
df['embarked'].mode()[0]
df["age"] = df["age"].fillna(df["age"].mean())
```

```python
df.isnull().sum()
```

```python
df["embarked"] = df['embarked'].fillna(df["embarked"].mode()[0])
df["embark_town"] = df['embark_town'].fillna(df["embark_town"].mode()[0])
```

```python
# Now verifying we have no missing values via heatmaps
df.isnull().sum()
sns.heatmap(df.isnull())
```

**Binning**:Binning (also called discretization) in data analysis means converting continuous numerical values into discrete groups (bins).

We take a range of numbers and divide them into intervals, then replace the exact values with the bin they belong to. 
1. Reduce noise in data
2. Simplify data (especially for analysis & visualization)
3. Convert continuous data into categorical data
4. Improve model performance in some ML algorithms
5. Make patterns easier to understand

```python
df.head()
df.age.min()
df.age.max()
sns.histplot(df["age"])
```

```python
# binning?cuttings of age column into 7 categories
# bins define range
# bins length is n+1, where n = number of bins.
# labels are the names given to each bin.
# labels length must be exactly one less than bins.
bins = [0,1,5,12,18,30,50,80]
labels  = ['Infants','Toddlers','Kids','Teens','Youngs','Middle aged','old']
# making class intervals
pd.cut(df["age"], bins = bins, labels = labels)
# len(bins)
# len(labels)
#   
```

```python
df.head()
# feature engineering for adding new column
df["binnned_age"] = pd.cut(df['age'], bins = bins, labels = labels)
df.head()
df['binnned_age'].value_counts()
# renamming a columns
df.rename(columns = {'binnned_age': 'age_groups'}, inplace = True)
# groupby analysis
df.groupby(["survived",'alive'])["age_groups"].value_counts()
# convert it into clean table
result  =(df.groupby('age_groups)["survived",'alive']
      .value_counts()
      .unstack(fill_value=0)
)

result


```
